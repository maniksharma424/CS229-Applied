```python
import pandas as pd
data = pd.read_csv("CarPrice_Assignment.csv")
data 
```

```python
data.tail(n=10)
```

```python
pd.set_option("display.max_columns", None)  # Show all columns
# dropped unnecessary cols 
data = data.drop(
    axis=1, labels=["CarName", "car_ID", "symboling", "boreratio"]
)
data

```

```python
# encoding categorical variables using one-hot encoding
data = pd.get_dummies(
    data,
    columns=[
        "fueltype",
        "aspiration",
        "doornumber",
        "carbody",
        "drivewheel",
        "enginetype",
        "cylindernumber",
        "fuelsystem",
        "enginelocation",
    ],
)
```

```python
sns.lmplot(
    data=data,
    x="citympg",
    y="price",
)
```

```python
# Defining Predictive Variables
X = data[
    [
        "wheelbase",
        "carlength",
        "carwidth",
        "carheight",
        "curbweight",
        "enginesize",
        "stroke",
        "compressionratio",
        "horsepower",
        "peakrpm",
        "citympg",
        "highwaympg",
        "fueltype_diesel",
        "fueltype_gas",
        "aspiration_std",
        "aspiration_turbo",
        "doornumber_four",
        "doornumber_two",
        "carbody_convertible",
        "carbody_hardtop",
        "carbody_hatchback",
        "carbody_sedan",
        "carbody_wagon",
        "drivewheel_4wd",
        "drivewheel_fwd",
        "drivewheel_rwd",
        "enginetype_dohc",
        "enginetype_dohcv",
        "enginetype_l",
        "enginetype_ohc",
        "enginetype_ohcf",
        "enginetype_ohcv",
        "enginetype_rotor",
        "cylindernumber_eight",
        "cylindernumber_five",
        "cylindernumber_four",
        "cylindernumber_six",
        "cylindernumber_three",
        "cylindernumber_twelve",
        "cylindernumber_two",
        "fuelsystem_1bbl",
        "fuelsystem_2bbl",
        "fuelsystem_4bbl",
        "fuelsystem_idi",
        "fuelsystem_mfi",
        "fuelsystem_mpfi",
        "fuelsystem_spdi",
        "fuelsystem_spfi",
        "enginelocation_front",
        "enginelocation_rear",
    ]
]

# Defining Target Variable
Y = data["price"]
X.describe(),Y.describe()
```

```python
# splitting test and train data 80/20 split 
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train, Y_train)
predictions = lm.predict(X_test )
# train the model run price predictions of test data 
predictions
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))

# Scatterplot: actual vs predicted
sns.scatterplot(x=Y_test, y=predictions, alpha=0.6)

# Add ideal line: y = x
plt.plot(
    [Y_test.min(), Y_test.max()],
    [Y_test.min(), Y_test.max()],
    color="red",
    linestyle="--",
    linewidth=2,
)

# Labels and title
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted Car Prices")
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(Y_test, predictions)
print(f"R² Score: {r2:.4f}")
# R² Score: 0.8820 considered good for regression models but can be made better as we have lot of features in the dataset that can be tuned .
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# deduce errors between prediction and actual values
print("mean absolute error ", mean_absolute_error(Y_test, predictions))
print("mean squared error ", mean_squared_error(Y_test, predictions))
# mean absolute error  2196.4982298892355 16% of the avg price (can be made better by feature engineering)
# mean squared error  9312213.038856803
```

```python
# residuals 
residuals = Y_test- predictions
sns.histplot(residuals, kde=True)
```
