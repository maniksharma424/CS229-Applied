```python
from sklearn.model_selection import train_test_split
```

```python
# Defining Predictive Variables
X = df[['Avg. Session Length' , 'Time on App' , 'Time on Website' , 'Length of Membership']]
#  Defining  Target Variables
Y = df['Yearly Amount Spent']
X,Y
```

```python
# make actual splits 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=100)
X_train,Y_train,X_test, Y_test
```

```python
# Train the model
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
# fit the data and run predict method  
lm.fit(X_train,Y_train)
predictions = lm.predict(X_test)
# plot predictions against test split to verify  
sns.scatterplot(x=predictions, y=Y_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
# deduce errors between prediction and actual values 
print("mean absolute error " , mean_absolute_error(Y_test,predictions))
print("mean squared error " , mean_squared_error(Y_test,predictions))
```

```python
# residuals are difference in predictions and actual values 
residuals = Y_test - predictions
residuals
# residual plot should be normally distributed as there should be no trend between errors because they are there because of the random noise and not because of something from our model 
sns.histplot(residuals, kde=True)
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
# Q-Q plot or prob plot: when it forms a straight line, it shows that residuals are normally distributed.
# It generates same number of inputs as of our residuals, and plots theoretical quantiles against actual residuals.
# If they coincide, it means residuals are normally distributed.
stats.probplot(residuals, dist="norm", plot=plt)
plt.show()
```
